# Data munging with regular expression

### Cyclic Voltametric data from CHI instrument (serial transported 'txt' file)



```python
cv = open('data/cvcurve1.txt', 'r')
print cv.read()
```

```python
import re
from pylab import *

find = r'([-]?[.0-9]{5}),\s([-]?[.0-9]{5}e[-]?\d+),\s[-]?[.0-9]{5}e[-]?\d+'
replace = r'\1,\2'

potential = []
current = []

for line in open('data/cvcurve1.txt', 'r'):
    if re.match(find, line):
        newline = re.sub(find, replace, line)
        a = newline.split(',')
        potential.append(a[0])
        current.append(a[1])
    else:
        pass
```

```python
figure(figsize=(8,6))

plot(potential, current, linewidth = 2.5)
xlabel('Potential (V) vs Ag/AgCl', fontsize = 20)
ylabel('Current (A)', fontsize = 20)
```

```python
import xlsxwriter, os

packed = zip(potential, current)

if 'output' in os.listdir(os.getcwd()):
    pass
else:
    os.mkdir('output')

workbook = xlsxwriter.Workbook('output/cv.xlsx')
worksheet = workbook.add_worksheet()

row = 0

for x, y in packed:
    worksheet.write(row, 0, x)
    worksheet.write(row, 1, y)
    row += 1

workbook.close()
```
