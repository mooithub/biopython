## **OSX에서 설치하기**

- homebrew 가 설치되어있으면 다음 명령어를 통해 쉽게 설치 가능하다.

> brew install mariadb

- 마리아db 서버를 시작하는 방법은 다음과 같다.
    
> mysql.server start

- 시스템이 부팅될 때 자동으로 실행되도록 하기 위해선 다음 명령을 실행한다.

> ln -sfv /usr/local/Cellar/mariadb/x.x.x(ver)/homebrew.mxcl.mariadb.plist ~/Library/LaunchAgents/

> launchctl load ~/Library/LaunchAgents/homebrew.mxcl.mariadb.plist
    
- 정지하는 방법은 다음과 같다
    
> launchctl unload ~/Library/LaunchAgents/homebrew.mxcl.mariadb.plist
    
- 마리아db 설치 후 신속히 테스트를 하기 위해선 다음과 같은 명령을 실행하자

> mysql -u root -p(password 사용하지 않는 경우에는 이 옵션 제거)

- 설치후 실행할것 (이유 아직 모름)

> unset TMPDIR
> mysql_install_db
    
- 보안설정
    
> mysql_secure_installation

- 마리아db 설치 후 비밀번호 설정 (위의 보안설정 수행시 자동으로 설정됨)

> mysqladmin -u root password 'new-password'
    
- 마리아db 재시작
    
> mysql reload

## **마리아db 사용자 관리**

### 사용자 권한

#### 전체에 적용되는 관리자권한

권한|설명
---|---
CREATE USER|CREATE USER 문을 사용해서 사용자를 생성하는 권한
FILE|LOAD DATA INFILE문과 LOAD_FILE() 함수를 사용하는 권한
PROCESS|SHOW PROCESSLIST 명령을 사용하는 권한
RELOAD|FLUSH 문을 사용하는 권한
REPLICATION CLIENT|SHOW MASTER STATUS 와 SHOW SLAVE STATUS 명령을 사용하는 권한
REPLICATION SLAVE|복제 마스터 서버에서 이루어진 업데이트 내용을 받을 권한
SHOW DATABASES|서버에서 데이터베이스의 목록을 볼 수 있는 권한
SHUTDOWN|mysqladmin shutdown 명령을 사용해서 서버를 중지할 수 있는 권한
SUPER|CHANGE MASTER TO... PURGE LOGS 명령과 전역변수를 세팅하고 <br>다른 사용자의 스레드를 강제로 중지시키는 등의 최고사용자가 사용하는<br> 구문을 사용하는 권한

#### 데이터베이스, 테이블, 칼럼 단위 권한

권한|설명
---|---
ALTER|인덱스와 테이블을 변경할 권한
ALTER ROUTINE|프로시저와 함수를 변경하거나 삭제하는 권한
CREATE|데이터베이스와 테이블을 생성하는 권한
CREATE ROUTINE|프로시저와 함수를 생성하는 권한
CREATE TEMPORARY TABLES|임시 테이블을 생성하는 권한
CREATE VIEW|뷰를 생성하는 권한
DELETE|테이블에서 행을 삭제하는 권한
DROP|전체 데이터베이스와 테이블을 삭제하는 권한
EVENT|이벤트 스케줄러에서 이벤트를 변경, 생성 및 삭제하는 권한
EXECUTE|함수와 프로시저를 실행하는 권한
INDEX|인덱스를 생성하거나 삭제하는 권한
INSERT|새로운 데이터를 테이블에 입력하는 권한
LOCK TABLE|테이블에 락을 걸거나 락을 다시 푸는 권한
SELECT|테이블에서 데이터를 읽는 권한
SHOW VIEW|SHOW CREATE VIEW 문을 사용하는 권한
TRIGGER|CREATE TRIGGER 와 DROP TRIGGER 문을 사용하는 권한
UPDATE|테이블에서 행을 수정하는 권한

- 컬럼 단위 권한은 테이블에서 개별 칼럼에 적용한다. 칼럼 단위 권환은 3가지(INSERT, UPDATE, SELECT) 가 있다.

#### 기타 부가적인 권한과 제한

권한|설명
---|---
USAGE|실제로는 아무것도 하지 않지만, 사용자에 대한 전역적인 설정을 변경하기 위해 사용할 수 있다.
ALL PRIVILEGES|사용자에게 사용 가능한 모든 권한을 부여하기 위해 사용할 수 있다.<br>GRANT OPTION 권한을 부여하지는 않는다. ALL로 줄여서 사용할 수 있다.
GRANT OPTION|가진 권한을 다른 사용자에게 주기 위한 권한을 사용자에게 부여한다.<br>GRANT 구문 마지막에 넣는다.


- 사용자에게 다음과 같은 몇가지 제한을 두는 것도 가능하다.

제한|설명
---|---
MAX_QUERIES_PER_HOUR|사용자별로 시간당 호출할 수 있는 SQL구문이나 쿼리의 수.<BR> 데이터를 변경하는 쿼리도 모두 포함한다.
MAX_UPDATES_PER_HOUR|사용자별로 시간당 호출할 수 있는 데이터를 변경하는 쿼리의 수.<BR> 조회성 쿼리는 제외된다.
MAX_CONNECTIONS_PER_HOUR|사용자가 시간당 접속을 시작할 수 있는 수
MAX_USER_CONNECTIONS|사용자가 데이터베이스 서버에 동시에 맺을 수 있는 연결 수.<BR> 만약 0으로 세팅되면 최대수인 MAX_CONNCETIONS으로  세팅한 것과 동일하다.<br> MAX_CONNECTIONS 또한 0이라서 동시에 연결할 수 있는 수에 제한이 없다.

## 사용자 생성

- 마리아db 에서 사용자 생성을 위해서는 다음과 같은 두 단계를 거친다
    1. CREATE USER 구문: 사용자 생성
    2. GRANT 명령: 사용자에게 권한을 부여
    
> CREATE USER '사용자명'@'호스트' IDENTIFIED BY '비밀번호';

- IDENTIFIED BY '비밀번호' 부분을 제거하면, 비밀번호 지정하지 않고 사용가능
- 모든 SQL 구문은 ';' 로 끝남
- 호스트는 사용자가 접속할 컴퓨터 또는 아이피주소 및 네트워크 등을 적어줄 수 있다.<br> 와일드카드 '%'을 사용하면 어디서든 접속할수 있도록 허용한다.

[CREATE USER문의 완전한 문서](https://mariadb.com/kb/en/create-user/)

## 퍼미션 부여

- 기본적으로 새로운 사용자는 어떤 권한도 갖지 않기 때문에 GRANT 명령을 이용해 부여해줄 필요가 있다.

> GRANT <권한> ON <데이터베이스> TO <사용자명>;

- 없는 사용자가 지정되었을 때는 비밀번호 없이 사용자가 생성된다. 이 경우 IDENTIFIED BY '비밀번호' 를 뒤에 기입한다.

> <예제>
> GRANT ALL ON \*.\* TO 'robert'@'%' WITH GRANT OPTION;

- 모든 권한 부여

> GRANT SELECT, INSERT, UPDATE, DELETE ON serv.* TO 'jeffrey'@'localhost';

- serv 데이터베이스에 읽고(SELECT) 쓰기(INSERT, UPDATE) 및 삭제(DELETE) 권한을 할당

> GRANT SELECT ON edu.staff TO 'david'@'localhost' WITH GRANT OPTION;

- edu 데이터베이스의 staff 테이블에 읽기 권한(SELECT)을 가지고 다른 사용자에게도 동일한 권한을 부여할 수 있도록<br>GRANT OPTION 권한을 줌

> GRANT ALL ON logan.* TO 'quentin'@'localhost' WITH MAX_QUERIES_PER_HOUR 100;

- logan 데이터베이스의 모든 권한을 부여하고 시간당 100개의 쿼리만 가능하도록 제한

[GRANT 구문의 모든 정보](http://mariadb.com/kb/en/grant/)

### 권한 추가와 삭제

- 권한 추가는 GRANT 를 동일하게 사용하되 권한을 제거하기 위해선 REVOKE 구문을 사용한다.

> REVOKE <권한> ON <데이터베이스> FROM <사용자명>;

[REVOKE 구문에 대한 모든 정보](https://mariadb.com/kb/en/revoke/)

### 권한부여 내용 보기

- 사용자에게 부여된 권한을 보기위해선 SHOW GRANT 구문을 사용한다.

> SHOW GRANTS FOR <사용자명>;

### 비밀번호 변경

- 사용자의 비밀번호를 변경하기 위해서는 SET PASSWORD 구문을 사용한다.

> SET PASSWORD FOR <사용자명> = PASSWORD('<비밀번호>');

[SET PASSWORD 구문의 모든 정보](https://mariadb.com/kb/en/set-password/)

### 사용자 제거

- 사용자를 완전히 제거하기 위해서 DROP USER 구문을 사용한다.<br>사용자가 제거되면 모든 권한이 자동으로 제거된다.

> DROP USER <사용자명>;

[DROP USER 구문의 모든 정보를 볼 수 있다](https://mariadb.com/kb/en/drop-user/)

### 권한 그룹

- 10.0.5 버전부터 개별 권한들을 묶어서 하나의 권한 그룹으로 만들고 이 권한그룹을 개별 사용자에게 부여하는 방식이 지원된다.
- 권한그룹의 생성은 다음과 같은 방법을 이용한다

> CREATE ROLE <권한그룹명>;

- 다음과 같은 사용이 가능하다.

> CREATE ROLE dba;
> CREATE ROLE developer WITH ADMIN [CURRENT_USER | CURRENT_ROLE | user | role]

- WITH ADMIN 절이 생략되면 기본적으로 'WITH ADMIN CURRENT_USER' 가 적용된다.
- CURRENT_USER 는 권한그룹을 생성하는 사용자, CURRENT_ROLE 은 그 사용자가 속해있는 그룹
- user 는 특정 사용자, role 은 특정 그룹을 지칭한다.
- 권한 그룹이 생성되면 다음과 같이 권한을 부여할 수 있다.

> GRANT ALL PRIVILEGES ON \*.\* TO dba;
> GRANT SELECT, INSERT, UPDATE, DELETE ON 'db1'.\* TO developer;

- 권한 그룹 부여는 다음과 같은 방법을 이용한다.

> GRANT <권한 그룹명> TO 'user'@'%';

- 현재 사용자에게 그룹을 부여하더라도 자동으로 인식되지 않는 문제점이 있다.

> SELECT CURRENT_ROLE;

- 위 명령어를 입력할 경우, 현재 ROLE 이 null로 확인된다.
- 이 경우 SET ROLE 명령어를 이용해 현재 세션에 권한그룹을 획득해야 한다.

> SET ROLE <권한 그룹>;

- 또한 현재 마리아 db 는 한 사용자가 2개 이상의 권한 그룹을 동시에 가질 수 없다는 단점이 있다.
- 설정된 권한그룹을 해지하기 위해서는 권한그룹을 NONE 으로 설정하면 된다.

> SET ROLE NONE;

## 데이터베이스의 생성 및 삭제

- 관리목적의 정보를 유지하는 system 데이터베이스
- 읽기만 가능한 통계용 performance_schema 데이터베이스

### 데이터베이스의 생성

> CREATE DATABASE <데이터베이스명>;

- 데이터베이스가 이미 존재하는 경우 발생하는 에러를 막기 위해 다음과 같은 옵션사용이 가능하다.

> CREATE DATABASE IF NOT EXISTS my_database;

[CREATE DATABASE 에 대한 모든 정보](https://mariadb.com/kb/en/create-database/)

### 데이터베이스의 삭제

> DROP DATABASE <데이터베이스명>;

- 생성과 마찬자기로 'IF EXISTS' 옵션을 넣으면 데이터베이스가 존제하지 않더라고 오류가 발생가지 않는다.

[DROP DATAVASE 에 대한 모든 정보](https://mariadb.com/kb/en/drop-database/)

## 테이블 생성, 변경, 삭제

- 테이블은 데이터베이스 내에 실질적으로 데이터값(레코드) 가 기록되어있는 곳이다.
- 엑셀의 스프레드시트 와 같은 것으로 생각하면 된다.

### 테이블 생성

- 테이블을 생성할 때는 'CREATE TABLE' 명령을 사용하며 다음과 같은 문법을 사용한다.
- 마찬가지로 IF NOT EXISTS 명령을 추가할 수 있다.

> CREATE TABLE 테이블명 (<칼럼 정의>);

- 중복 데이터나 거의 사용하지 않는 부가적인 칼럼을 생성하지 않기 위해 테이블을 새롭게 정의해서 서로 연결해 관계를 정의해주는데<br>이러한 테이블 정의를 재정의 하는 것을 정규화라고 부른다.

[정규화 과정에 대한 좀 더 깊이있는 정보](https://mariadb.com/kb/en/recap-the-relational-model)

- <칼럼정의> 부분은 다음과 같은 기본 형태를 가진다.

    <칼럼명> <데이터타입>
        [NOT NULL | NULL]
        [DEFAULT <기본값>]
        [AUTO_INCREMENT]
        [UNIQUE [KEY] | [PRIMARY] KEY]
        [COMMENT '주석']

- '<>' 안의 부분을 태워야 하며, '[]' 안은 옵션이고 '|'(파이프문자) 를 이용해 선택 가능한 값을 알려준다.
- NOT NULL 또는 NULL 또는 지정하지 않을 수 있으며, NOT NULL로 지정되어있을 때는 그 칼럼은 반드시 값을 가져야 한다.
- <데이터타입> 은 INTEGER, FLOAT (숫자형), CHAR, TEXT, VARCHAR (문자형), DATE, TIME, DATETIME (날짜/시각) 등이 있다.

[데이터 타입의 목록](https://mariadb.com/kb/en/datatypes/)

> (예제)

    CREATE TABLE employees (
        id INT NOT NULL AUTO_INCREMENT PRIMARY KEY,
        surname VARCHAR(100),
        givenname VARCHAR(100),
        pref_name VARCHAR(50),
        birthday DATE
    );
      
[CREATE TABLE 명령에 대한 모든 정보](https://mariadb.com/kb/en/create-talbe/)

> SHOW CREATE TABLE <테이블명>\G(또는 ;)

- 위 명령을 수행하면 테이블을 다시 만들기 위한 명령을 얻을 수 있다.

[SHOW CREATE TABLE 명령에 대한 모든 정보](https://mariadb.com/kb/en/show-create-table/)

> DESCRIBE <테이블명>;

- 위 명령을 수행하면 테이블의 구조를 한눈에 볼 수 있다.

[DESCRIBE 명령에 대한 모든 정보](https://mariadb.com/kb/en/describe/)

### 테이블 변경

- 테이블을 정의하더라도 구조를 변경할 필요가 생긴다. 이 경우 ALTER TABLE 명령을 사용해 구조를 변경한다.

> ALTER TABLE 테이블명 <변경 정의>[, 변경 정의] ...;

- <변경 정의> 속성은 ADD, MODIFT, DROP 을 지정해서 각각 칼럼을 추가, 변경, 삭제 할 수 있다.
- 한개의 ALTER TABLE 명령에서 여러개의 변경정의를 둘 때는 각각을 콤마로 구분해서 나열한다.

### 칼럼 추가

- ALTER TABLE 명령의 변경정의 속성에 다음의 형태를 사용해 칼럼을 추가할 수 있다.

    ADD <칼럼명> <칼럼정의> [FIRST | AFTER <칼럼명>]

- FIRST 와 AFTER 부분은 설정하지 않아도 무방하다. FIRST 옵션은 행의 첫번째 칼럼에 새로운 칼럼을 추가하며
- AFTER 옵션은 명시한 칼럼 뒤에 새로운 칼럼을 추가한다.

> ALTER TABLE employees ADD username varchar(20) AFTER pref_name;

### 칼럼 변경

    MODIFY <칼럼명> <칼럼 정의>
    
> ALTER TABLE employees MODIFY pref_name varchar(25);

### 칼럼 삭제

    DROP <칼럼명>
    
> ALTER TABLE employees DROP username;

[ALTER TABLE 명령에 대한 모든 정보](https://mariadb.com/kb/en/alter-table/)

### 테이블 삭제

> DROP TABLE <테이블명>

- 삭제하려는 테이블이 없는 경우 오류가 발생하므로, 'IF EXISTS' 옵션을 사용하는 것이 유용하다.

[DROP TABLE 명령에 대한 모든 정보](https://mariadb.com/kb/en/drop-table/)

## 테이터 조회, 입력, 수정 및 삭제

### 데이터 입력

> INSERT [INTO] <테이블명> [(<칼럼명>[, 칼럼명,...])]<br>
{VALUES } VALUE}<br>
({표현식 | DEFAULT}, ...)[,(...),...];

- CREATE TABLE 처럼 '<>' 안의 내용은 다른 값으로 치환 가능하다.
- '[]' 안의 부분은 사용하지 않아도 되며, '|' 사용해서 선택 가능한 값을 알려준다.
- '{}' 는 필수값을 나타내지만, 여기서는 사용할 수 있는 키워드를 선택한다.
- 표현식은 칼럼에 넣을 값을 나열한다.

    INSERT INTO employees (surname, givenname) VALUES
        ("Taylor", "John"),
        ("Woodruff", "Wilford"),
        ("Snow", "Lorenzo");
    
    INSERT INTO employees (pref_name, givenname, surname, birthday)
        VALUES ("George", "George Albert", "Smith", "1940-04-04");
    
    INSERT employees (surname) VALUE ("MaKay");
    INSERT INTO employees VALUES
        (NULL, "Kimball", "Spencer", NULL, NULL);

[INSERT 명령에 대한 모든 정보](https://mariadb.com/kb/en/insert/)

### 데이터 수정

    UPDATE <테이블명><br>
        SET 칼럼명={표현식|DEFAULT}
        [, 칼럼명={표현식|DEFAULT}] ...
        WHERE <조회조건>];

- INSERT 명령과는 달리 데이터를 수정할 때 각각의 칼럼명 뒤에 입력한 데이터를 지정한다.
- WHERE 조건을 사용하는 것도 또 다른 차이점이다. - 변경하고자 하는 테이블의 칼럼을 지정하기 위한 중요한 옵션
- WHERE 조건을 생략하면 UPDATE 구문은 선택한 칼럼의 모든 데이터를 수정한다.

    UPDATE empoyees SET
        pref_name = "John", birthday = "1958-11-01"
        WHERE surname = "Taylor" AND givenname = "John";
        
    UPDATE employees SET
        pref_name = "Will", birthday = "1957-03-01"
        WHERE surname = "Woodruff";
        
    UPDATE employees SET
        birthday = "1964-04-03"
        WHERE surname = "Snow";
        
- 위 명령어를 실행하면 다음과 같은 결과가 출력된다.

    Query OK, 1 row affected (0.03 sec)
    Rows matched: 1 Changed: 1 Warnings: 0

- 위 두줄은 WHERE 조건에 일치하는 행이 몇 개 인지 알려준다.

[UPDATE 명령에 대한 모든 정보](https://mariadb.com/kb/en/update/)

### 데이터 삭제

> DELETE FROM <테이블명> [WHERE <조회조건>];

- WHERE 조건을 사용하지 않으면 테이블의 모든 행을 삭제한다.

    DELETE FROM employees
    WHERE givenname="Spencer" AND surname="Kimball";

[DELETE 명령에 대한 모든 정보](https://mariadb.com/kb/en/delete/)

### 데이터 읽기

    SELECT <대상칼럼> FROM <테이블명>
        [WHERE <검색조건>]
        [ORDER BY <칼럼명>];

- 대상 칼럼에 가져올 칼럼을 모두 적는다. '\*'도 사용 가능하다.

    SELECT * FROM emplyees;
    
    SELECT pref_name, surname FROM employees;
    
    SELECT * FROM employees
        WHERE birthday > '1960-01-01';
    
- '>' 부호는 비교연산자로 기준값보다 큰 것을 의미한다.

[비교연산자 목록](https://mariadb.com/kb/en/comparison-operator/)

- 직원 목록을 surname 칼럼을 기준으로 정렬한다면 ORDER BY 절을 사용한다.

    SELECT * FROM employees ORDER BY surname;
    
- McK 문자로 시작하는 이름을 가진 직원을 찾길 원하는 경우 WHERE 조건을 다음과 같이 이용해 검색 가능하다.

    SELECT * FROM employees
        WHERE surname LIKE "McK%";

[SELECT 명령에 대한 모든 정보](https://mariadb.com/kb/en/select/)

- SELECT 명령은 여러 개의 테이블에서 데이터를 모으기 위해 조인을 사용하면 더욱 강력해 진다.
- 예를 들어 employee 테이블과 관련된 address테이블을 가지고 있다면, 어떤 도시에 살고 생일이 한 달 이내에 들어있는<br>모든 직원의 이름과 주소를 조회해서 그 도시의 식당에서 식사를 할 수 있는 선물을 줄 수 있다.

[JOIN 에 대한 모든 정보](https://mariadb.com/kb/en/join/)
